```python
import itertools
```

## Layer list in LayerTree element

```python
layer_template = """<Layer id="1995-12-12 {time}" name="1995-12-12 {time}">
    <Layer id="Nelson 1995-12-12 {time}" name="Nelson"/>
    <Layer id="Collingwood 1995-12-12 {time}" name="Collingwood"/>
</Layer>"""
```

```python
print(layer_template.format(time="05:00"))
```

```python
hours = range(5, 12)
mins = [0, 15, 30, 45]

for h, m in itertools.product(hours, mins):
    time_str = f"{h:02}:{m:02}"
    print(layer_template.format(time=time_str))
```

## Layers in Content section

```python
content_template = """<IfLayerOn layerId="1995-12-12 {time}">
    <IfLayerOn layerId="Nelson 1995-12-12 {time}">
        <Vector dataset="15min_split/Nelson_Lines_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='PEN(c:#bd1b44,w:5px)'/>
        <Vector dataset="15min_split/Nelson_10min_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='SYMBOL(c:#bd1b44,s:2,id:"ogr-sym-3")'>
            <LogicalStructure displayLayerName="Nelson" fieldToDisplay="time"/>
        </Vector>
        <VectorLabel dataset="15min_split/Nelson_Hourly_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='LABEL(t:{time_str},c:#bd1b44,s:24pt,p:4,dx:7mm,bo:1)'/>
        <VectorLabel dataset="Nelson_FirstPoint.geojson" layer="Nelson_FirstPoint" georeferencingId="georeferenced" ogrStyleString='LABEL(t:"Nelson",c:#bd1b44,s:24pt,p:2,dy:10mm,bo:1)'/>
    </IfLayerOn>

    <IfLayerOn layerId="Collingwood 1995-12-12 {time}">
        <Vector dataset="15min_split/Collingwood_Lines_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='PEN(c:#0d0d8e,w:5px)'/>
        <Vector dataset="15min_split/Collingwood_10min_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='SYMBOL(c:#0d0d8e,s:2,id:"ogr-sym-3")'>
            <LogicalStructure displayLayerName="Collingwood" fieldToDisplay="time"/>
        </Vector>
        <VectorLabel dataset="15min_split/Collingwood_Hourly_Split/Nearest15min_{timenocolon}.geojson" layer="Nearest15min_{timenocolon}" georeferencingId="georeferenced" ogrStyleString='LABEL(t:{time_str},c:#0d0d8e,s:24pt,p:4,dx:7mm,bo:1)'/>
        <VectorLabel dataset="Collingwood_FirstPoint.geojson" layer="Collingwood_FirstPoint" georeferencingId="georeferenced" ogrStyleString='LABEL(t:"Collingwood",c:#0d0d8e,s:24pt,p:2,dy:-10mm,bo:1)'/>
    </IfLayerOn>
</IfLayerOn>"""
```

```python
hours = range(5, 12)
mins = [0, 15, 30, 45]

for h, m in itertools.product(hours, mins):
    time_str = f"{h:02}:{m:02}"
    time_str_no_colon = f"{h:02}{m:02}"
    print(content_template.format(time=time_str, timenocolon=time_str_no_colon, time_str='{time_str}'))
```
